```python
X=data.drop('charges', axis=1)
y=data.charges
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
continuous_cols = ['age', 'bmi', 'children']
binary_cols = ['sex','smoker','northeast','northwest','southeast','southwest']
from sklearn.preprocessing import StandardScaler
# scale the continuous features
scaler = StandardScaler()
X_train_cont = scaler.fit_transform(X_train[continuous_cols])
X_test_cont = scaler.transform(X_test[continuous_cols])
```

```python
data
```

```python
X_train_scaled = pd.DataFrame(data=X_train_cont, columns=continuous_cols)
X_train_scaled[binary_cols] = X_train[binary_cols].reset_index(drop=True)
X_test_scaled = pd.DataFrame(data=X_test_cont, columns=continuous_cols)
X_test_scaled[binary_cols] = X_test[binary_cols].reset_index(drop=True)
```

```python
X_train_scaled
```

```python
X_train_scaled[binary_cols]
```

```python
from sklearn.linear_model import LinearRegression

# prepare the training data
X_train = X_train_scaled
y_train = y_train

# prepare the test data
X_test = X_test_scaled
y_test = y_test

# train the model
lr = LinearRegression()
lr.fit(X_train, y_train)

# predict on the test data
y_pred_lr= lr.predict(X_test)

# evaluate the model
from sklearn.metrics import r2_score, mean_squared_error

r2 = r2_score(y_test, y_pred_lr)
mse = mean_squared_error(y_test, y_pred_lr)
rmse = np.sqrt(mse)


print("R-squared:", r2)
print("Mean squared error:", mse)
print("Root mean squared error:", rmse)
lr_r2 = r2_score(y_test, y_pred_lr)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score, mean_squared_error

# instantiate the model
ridge = Ridge(alpha=1)

# fit the model on the training data
ridge.fit(X_train, y_train)

# make predictions on the test data
y_pred_rgd = ridge.predict(X_test)

# calculate evaluation metrics
print('R-squared:', r2_score(y_test, y_pred_rgd))
print('Mean squared error:', mean_squared_error(y_test, y_pred_rgd))
print('Root mean squared error:', np.sqrt(mean_squared_error(y_test, y_pred_rgd)))
ridge_r2 = r2_score(y_test, y_pred_rgd)
```

```python
from sklearn.tree import DecisionTreeRegressor

# Create the model
dt_reg = DecisionTreeRegressor(random_state=42)

# Fit the model on the training data
dt_reg.fit(X_train, y_train)

# Predict on the test data
y_pred_dt= dt_reg.predict(X_test)

# Evaluate the model
from sklearn.metrics import r2_score, mean_squared_error
print("R-squared:", r2_score(y_test, y_pred_dt))
print("Mean squared error:", mean_squared_error(y_test, y_pred_dt))
print("Root mean squared error:", np.sqrt(mean_squared_error(y_test, y_pred_dt)))
tree_r2 = r2_score(y_test, y_pred_dt)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Create the model
rf_reg = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model on the training data
rf_reg.fit(X_train, y_train)

# Predict on the test data
y_pred_rf= rf_reg.predict(X_test)

# Evaluate the model
from sklearn.metrics import r2_score, mean_squared_error
print("R-squared:", r2_score(y_test, y_pred_rf))
print("Mean squared error:", mean_squared_error(y_test, y_pred_rf))
print("Root mean squared error:", np.sqrt(mean_squared_error(y_test, y_pred_rf)))
rf_r2 = r2_score(y_test, y_pred_rf)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# Create the model
gb_reg = GradientBoostingRegressor(n_estimators=100, random_state=42)

# Fit the model on the training data
gb_reg.fit(X_train, y_train)

# Predict on the test data
y_pred_gdb= gb_reg.predict(X_test)

# Evaluate the model
from sklearn.metrics import r2_score, mean_squared_error
print("R-squared:", r2_score(y_test, y_pred_gdb))
print("Mean squared error:", mean_squared_error(y_test, y_pred_gdb))
print("Root mean squared error:", np.sqrt(mean_squared_error(y_test, y_pred_gdb)))
gdb_r2 = r2_score(y_test, y_pred_gdb)
```

```python
from sklearn.svm import SVR

# Create the model
svr_reg = SVR(kernel='linear')

# Fit the model on the training data
svr_reg.fit(X_train, y_train)

# Predict on the test data
y_pred_svr= svr_reg.predict(X_test)

# Evaluate the model
from sklearn.metrics import r2_score, mean_squared_error
print("R-squared:", r2_score(y_test, y_pred_svr))
print("Mean squared error:", mean_squared_error(y_test, y_pred_svr))
print("Root mean squared error:", np.sqrt(mean_squared_error(y_test, y_pred_svr)))
svr_r2 = r2_score(y_test, y_pred_svr)
```

```python
from sklearn.neighbors import KNeighborsRegressor

# Create the model
knn_reg = KNeighborsRegressor()

# Fit the model on the training data
knn_reg.fit(X_train, y_train)

# Predict on the test data
y_pred_knn= knn_reg.predict(X_test)

# Evaluate the model
from sklearn.metrics import r2_score, mean_squared_error
print("R-squared:", r2_score(y_test, y_pred_knn))
print("Mean squared error:", mean_squared_error(y_test, y_pred_knn))
print("Root mean squared error:", np.sqrt(mean_squared_error(y_test, y_pred_knn)))
knn_r2 = r2_score(y_test, y_pred_knn)
```

```python
# AdaBoost regression
from sklearn.ensemble import AdaBoostRegressor

# create the model
ada = AdaBoostRegressor()

# fit the model on the training data
ada.fit(X_train, y_train)

# predict on the test data
y_pred_adb= ada.predict(X_test)

# evaluate the model
print("R-squared:", r2_score(y_test, y_pred_adb))
print("Mean squared error:", mean_squared_error(y_test, y_pred_adb))
print("Root mean squared error:", np.sqrt(mean_squared_error(y_test, y_pred_adb)))
adb_r2 = r2_score(y_test, y_pred_adb)
```

```python
models = pd.DataFrame([lr_r2,ridge_r2,tree_r2, rf_r2,gdb_r2,svr_r2, knn_r2, adb_r2],
             index=['Linear Regression', 'Ridge Regressor','Decision Tree','Random Forest','Gradient Boosting','SVR','KNN', 'AdaBoost'],columns=['R2_Score'])
models
```
